```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import os

# Initialize a dictionary to store each DataFrame
dataframes = {}

# Load each uploaded CSV into a DataFrame
for file_name in uploaded.keys():
    # Load CSV file into DataFrame
    df = pd.read_csv(file_name)
    dataframes[file_name] = df

# Combine all DataFrames by concatenating rows (or use merge logic if required)
combined_df = pd.concat(dataframes.values(), ignore_index=True)

# Save the combined DataFrame as a Pickle file
combined_df.to_pickle("combined_data.pkl")

# Optional: Save individual DataFrames in multiple sheets in an Excel file
with pd.ExcelWriter("combined_data.xlsx") as writer:
    for sheet_name, df in dataframes.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

# Download the files
from google.colab import files
files.download("combined_data.pkl")
files.download("combined_data.xlsx")
```
