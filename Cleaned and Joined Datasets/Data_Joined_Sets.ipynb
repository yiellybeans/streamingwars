```python
turnover = pd.read_csv(wkg_path + r"\SVOD Content Turnover (2020-2021).csv", header=3)
```

```python
turnover.head(10)
```

```python
content_titles = pd.read_csv(wkg_path + r"\content_titles_june_2017_2021.csv")
```

```python
content_titles.head(5)
```

```python
content_titles_june_2020 = content_titles[content_titles['Date'] == 'June 2020']

content_titles_june_2020.head(5)
```

```python
content_titles_june_2020 = content_titles_june_2020.T

content_titles_june_2020.head(10)
```

Drop the first two rows

```python

content_titles_june_2020 = content_titles_june_2020[2:]
```

```python
content_titles_june_2020 = content_titles_june_2020.rename(columns={3: 'Number of contents June 2020'})
```

```python
content_titles_june_2020.head(5)
```

Reset the index

```python

content_titles_june_2020 = content_titles_june_2020.reset_index()
```

```python
content_titles_june_2020.head(5)

content_titles_june_2020 = content_titles_june_2020.rename(columns={'index': 'Platform'})
```

```python
content_titles_june_2020.set_index('Platform', inplace=True)
```

```python
turnover.set_index('Platform', inplace=True)
```

Merge the two dataframes

```python

turnover = pd.merge(turnover, content_titles_june_2020, on='Platform')
```

```python
turnover.head(5)
```

```python
turnover['turnover_rate_i'] = turnover['Newly Added'] + turnover['Recently Removed']

turnover['turnover_rate %'] = turnover['turnover_rate_i'] / turnover['Number of contents June 2020'] *100
```

Delete the column 'turnover_rate_i'

```python

turnover = turnover.drop(columns=['turnover_rate_i'])

turnover.head(5)
```

Save as a CSV file

```python
turnover.to_csv(wkg_path + r"\turnover_rate_2020_2021.csv", index=False)
```
